### Пункт 1 
Изобразить в первом квадранте координатной плоскости свои инициалы (первые буквы имени и фамилии), соединяя точки отрезками прямых линий.

### Пункт 2 
Записать соединяемые точки в виде списков строк.

```python
LETTER_1 = [
    [1, 3],
    [1, 5],
    [2, 5],
    [3, 4],
    [2, 3],
    [1, 3],
    [1, 1],
    [2, 1],
    [3, 2],
    [2, 3],
]
LETTER_2 = [
    [6, 1], 
    [4, 3], 
    [6, 5], 
    [4, 3], 
    [4, 5], 
    [4, 1]
]
```

### Пункт 3 
Преобразовать списки строк в матрицы numpy. Эти матрицы будут исходными для следующих задач.

```python
letter_1 = Matrix.from_list(LETTER_1)
letter_2 = Matrix.from_list(LETTER_2)
```

### Пункт 4 
Вывести матрицы в виде рисунка. Проверить, что получились правильные буквы.

```python
plot_2d_letter(letter_1)
plot_2d_letter(letter_2)
```

### Пункт 5 
Умножить эти матрицы на такую матрицу, которая отражает плоскость относительно оси Ox. Вывести результат в виде рисунка.


```python
letter_1_flipped = letter_1.flip2d(0)
letter_2_flipped = letter_2.flip2d(0)

plot_2d_letter(letter_1)
plot_2d_letter(letter_2)

plot_2d_letter(letter_1_flipped)
plot_2d_letter(letter_2_flipped)
```

### Пункт 6 
Записать функцию, вычисляющую матрицу поворота на заданный угол. Умножить исходные матрицы на матрицы поворота на 120 и 240 градусов. Затем увеличить рисунок по каждой из осей в 5 раз и повернуть 5 раз на 60 градусов. Вывести все полученные матрицы в виде рисунка.


```python
plot_2d_letter(letter_1)
plot_2d_letter(letter_2)

plot_2d_letter(letter_1.rotate2d(120))
plot_2d_letter(letter_2.rotate2d(120))

plot_2d_letter(letter_1.rotate2d(240))
plot_2d_letter(letter_2.rotate2d(240))
```

```python
letter_1_scaled = letter_1.scale(5)
letter_2_scaled = letter_2.scale(5)

for k in range(0, 6):
    rotation_angle = 60 * k
    plot_2d_letter(letter_1_scaled.rotate2d(rotation_angle))
    plot_2d_letter(letter_2_scaled.rotate2d(rotation_angle))
```

### Пункт 7
Умножить исходные матрицы на матрицу 
$\left[
    \begin{matrix}
        −1 & 1 \\
        4 & 0
    \end{matrix}
\right]$. 
Вывести результат в виде
рисунка. Найти обратную матрицу, сделать проверку и умножить результат
преобразований на нее. Вывести результат обратного преобразования в виде рисунка.

```python
mult_matrix = Matrix.from_list([[-1, 1], [4, 0]])

letter_1_mult = letter_1 @ mult_matrix
letter_2_mult = letter_2 @ mult_matrix

plot_2d_letter(letter_1)
plot_2d_letter(letter_2)
plot_2d_letter(letter_1_mult)
plot_2d_letter(letter_2_mult)
```

```python
mult_matrix_inv = mult_matrix.inv()

letter_1_test_inv = letter_1 @ mult_matrix @ mult_matrix_inv
letter_2_test_inv = letter_2 @ mult_matrix @ mult_matrix_inv

plot_2d_letter(letter_1_test_inv)
plot_2d_letter(letter_2_test_inv)
```
